## Shortcut Keys

- Press `SHIFT + ENTER` to run a cell
- To convert Python cell to markdown `ESC + M`
- To convert markdown cell to python `ESC + Y`
- To Create a cell below press `ESC + B`
- To Create a cell above press `ESC + A`
- To delete a cell press `ESC + double D ("D" should be pressed 2-times)`




## Conditional Statements

- if internet_availible --> The class will be conducted
- if internet_not_availible --> Class will not be conducted


```python
internet = True
```

```python
if internet:
    print("Class will be conducted")
    print("Go to the institute")
    print("Atend the class")
else:
    print("Class will not be conducted")   
```
