```python
numbers = [1,2,2,3,3,3,4,4,4,4]

freq = {}
for num in numbers:
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1

print(freq)
```

```python
sentence = input("Enter a sentence: ")
words= sentence.split()
freq = {}
for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1
print(freq)       
```

```python
marks = {"Math": 80,"English": 70, "Physics": 90}
total_marks = sum(marks.values())
average_marks = total_marks/len(marks)
print(f"Total Marks: {total_marks}")
print(f"Average Marks: {average_marks}")
```

```python
students = {
    "Ali": 85,
    "Sara":92,
    "Ahmed": 78,
    "Zara":95
}
highest_score_student = max(students,key=students.get)
print(f"Student with highest score is {highest_score_student} with score {students[highest_score_student]}")
```
